```python
# This opens an upload button in Colab - choose a product photo, logo, or brand image
uploaded = files.upload()

# Get the filename automatically
image_path = list(uploaded.keys())[0]
print(f"Uploaded: {image_path}")
```

```python
color_thief = ColorThief(image_path)

# Get the single most dominant color
dominant_color = color_thief.get_color(quality=1)

# Get a palette of the top colors (e.g. 6 colors)
palette = color_thief.get_palette(color_count=6, quality=1)

print("Dominant Color (RGB):", dominant_color)
print("\nFull Palette (RGB):")
for i, color in enumerate(palette):
    print(f"  Color {i+1}: {color}")
```

```python
def rgb_to_hex(rgb):
    return '#{:02x}{:02x}{:02x}'.format(rgb[0], rgb[1], rgb[2])

hex_palette = [rgb_to_hex(color) for color in palette]

print("Palette in HEX:")
for i, hex_code in enumerate(hex_palette):
    print(f"  Color {i+1}: {hex_code}")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Show original image
img = Image.open(image_path)
axes[0].imshow(img)
axes[0].axis('off')
axes[0].set_title("Original Image")

# Show color swatches
swatch = [[color for color in palette]]
axes[1].imshow(swatch, aspect='auto')
axes[1].set_xticks(range(len(palette)))
axes[1].set_xticklabels(hex_palette, rotation=45, ha='right')
axes[1].set_yticks([])
axes[1].set_title("Extracted Palette")

plt.tight_layout()
plt.savefig("brand_palette_output.png", dpi=150)
plt.show()
```
